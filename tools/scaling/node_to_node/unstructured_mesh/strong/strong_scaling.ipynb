```python
import numpy as np
import pandas as pd
from IPython.display import display, HTML
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# LLNL Dane - 112 Intel Sapphire Rapids cores/node, 105MB cache/CPU, 2.28GB DDR5 per core
dane_nodes = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]
dane_ideal = [21.53, 21.53/2, 21.53/4, 21.53/8, 21.53/16, 21.53/32, 21.53/64, 21.53/128, 21.53/256, 21.53/512]
dane_10292024 = [21.53, 9.92, 4.76, 2.18, 0.99, 0.51, 0.26, 0.14, 0.09, 0.07]
```

```python
plt.style.use('seaborn-poster')
plt.suptitle("OpenSn Unstructured Mesh Node-to-Node Strong Scaling", fontsize=22)
plt.title("~256 cells/rank, 448 angles, 64 groups, single-angle aggregation", fontsize=14)
plt.grid(linestyle='--')
plt.xlabel('Compute Nodes [log2]');
plt.xscale('log', base=2)
plt.ylabel('Avg. Sweep Time (s) [log2]');
plt.yscale('log', base=2)

# LLNL Dane
plt.plot(dane_nodes, dane_ideal, linestyle='--', color='#878787', label='LLNL Dane (64 rpn) - Ideal')
plt.plot(dane_nodes, dane_10292024, marker='s', markersize=8, markerfacecolor='#252525', color='#252525', label='LLNL Dane (64 rpn) - 10/29/2024')

plt.legend()
#plt.savefig("opensn-strong-scaling.png")
```
